# Hands-on Lab


## Visualizing distribution of data


### Histograms


Plot a histogram of `ConvertedComp.`


```python
import matplotlib.pyplot as plt

QUERY = """
SELECT ConvertedComp
FROM master
"""

df_converted_comp = pd.read_sql_query(QUERY, conn)

# Plotting a histogram
plt.figure(figsize=(10, 6))
plt.hist(df_converted_comp['ConvertedComp'], bins=30, color='skyblue', edgecolor='black')
plt.xlabel('Converted Compensation')
plt.ylabel('Frequency')
plt.title('Histogram of Converted Compensation')
plt.grid(True)
plt.show()
```

### Box Plots


Plot a box plot of `Age.`


```python
import seaborn as sns

QUERY_AGE = """
SELECT Age
FROM master
"""

df_age = pd.read_sql_query(QUERY_AGE, conn)

# Plotting a box plot using seaborn
plt.figure(figsize=(8, 6))
sns.boxplot(x=df_age['Age'], color='lightblue')
plt.xlabel('Age')
plt.title('Box Plot of Age')
plt.grid(True)
plt.show()

```

## Visualizing relationships in data


### Scatter Plots


Create a scatter plot of `Age` and `WorkWeekHrs.`


```python
QUERY_AGE_WORKWEEKHRS = """
SELECT Age, WorkWeekHrs
FROM master
"""

df_age_workweekhrs = pd.read_sql_query(QUERY_AGE_WORKWEEKHRS, conn)

# Plotting a scatter plot
plt.figure(figsize=(10, 6))
plt.scatter(df_age_workweekhrs['Age'], df_age_workweekhrs['WorkWeekHrs'], color='orange', alpha=0.6)
plt.xlabel('Age')
plt.ylabel('Work Week Hours')
plt.title('Scatter Plot of Age vs Work Week Hours')
plt.grid(True)
plt.show()

```

### Bubble Plots


Create a bubble plot of `WorkWeekHrs` and `CodeRevHrs`, use `Age` column as bubble size.


```python
QUERY_WORK_CODE_AGE = """
SELECT WorkWeekHrs, CodeRevHrs, Age
FROM master
"""

df_work_code_age = pd.read_sql_query(QUERY_WORK_CODE_AGE, conn)

# Plotting a bubble plot
plt.figure(figsize=(12, 8))
bubble = plt.scatter(
    df_work_code_age['WorkWeekHrs'],
    df_work_code_age['CodeRevHrs'],
    s=df_work_code_age['Age']*10,  # Scale the bubble size based on Age
    alpha=0.5, cmap='viridis'
)
plt.colorbar(bubble, label='Age (Bubble Size)')
plt.xlabel('Work Week Hours')
plt.ylabel('Code Review Hours')
plt.title('Bubble Plot of Work Week Hours vs Code Review Hours')
plt.grid(True)
plt.show()
```

## Visualizing composition of data


### Pie Charts


Create a pie chart of the top 5 databases that respondents wish to learn next year. Label the pie chart with database names. Display percentages of each database on the pie chart.


```python
QUERY_TOP_DATABASES = """
SELECT DatabaseDesireNextYear, COUNT(DatabaseDesireNextYear) as count
FROM DatabaseDesireNextYear
WHERE DatabaseDesireNextYear IS NOT NULL
GROUP BY DatabaseDesireNextYear
ORDER BY count DESC
LIMIT 5
"""

df_top_databases = pd.read_sql_query(QUERY_TOP_DATABASES, conn)

# Plotting a pie chart
plt.figure(figsize=(8, 8))
plt.pie(df_top_databases['count'], labels=df_top_databases['DatabaseDesireNextYear'], autopct='%1.1f%%', startangle=140)
plt.title('Top 5 Databases Respondents Wish to Learn Next Year')
plt.show()

```

### Stacked Charts


Create a stacked chart of median `WorkWeekHrs` and `CodeRevHrs` for the age group 30 to 35.


```python
QUERY = """
SELECT WorkWeekHrs, CodeRevHrs, Age
FROM master
"""
df_age = pd.read_sql_query(QUERY,conn) 
#print(df_age.shape)

# group respondents by age and apply median() function 
df_age = df_age.groupby('Age', axis=0).median()
#df_age = df_age[30:35]
#print(df_age.shape)

df_age[30:35].plot(kind='bar', figsize=(10, 6), stacked=True)

plt.xlabel('Age') # add to x-label to the plot
plt.ylabel('Hours') # add y-label to the plot
plt.title('Median Hours by Age') # add title to the plot

plt.show()

```

## Visualizing comparison of data


### Line Chart


Plot the median `ConvertedComp` for all ages from 45 to 60.


```python
QUERY_MEDIAN_COMP = """
SELECT Age, ConvertedComp
FROM master
WHERE Age BETWEEN 25 AND 30
"""

df_median_comp = pd.read_sql_query(QUERY_MEDIAN_COMP, conn)

df_median_comp_grouped = df_median_comp.groupby('Age').median().reset_index()

# Plotting the line chart
plt.figure(figsize=(12, 6))
plt.plot(df_median_comp_grouped['Age'], df_median_comp_grouped['ConvertedComp'], marker='o', linestyle='-', color='b')
plt.xlabel('Age')
plt.ylabel('Median Converted Compensation')
plt.title('Median Converted Compensation for Ages 25 to 30')
plt.grid(True)
plt.show()

```

### Bar Chart


Create a horizontal bar chart using column `MainBranch.`


```python
QUERY_MAINBRANCH = """
SELECT MainBranch, COUNT(*) as count
FROM master
GROUP BY MainBranch
"""

df_mainbranch = pd.read_sql_query(QUERY_MAINBRANCH, conn)

# Plotting the horizontal bar chart
plt.figure(figsize=(10, 8))
plt.barh(df_mainbranch['MainBranch'], df_mainbranch['count'], color='skyblue')
plt.xlabel('Count')
plt.ylabel('Main Branch')
plt.title('Count of Respondents by Main Branch')
plt.grid(True)
plt.show()

```
